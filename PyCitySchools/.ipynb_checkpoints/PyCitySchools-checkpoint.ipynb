# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending per student actually (\$645-675) underperformed compared to schools with smaller budgets (<\$585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

```python
# Dependencies and Setup
import pandas as pd
```

```python
# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
```

```python
school_data = pd.read_csv(school_data_to_load)
#school_data
```

```python
# Read School and Student Data File and store into Pandas Data Frames
#school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
#student_data.head()
```

```python
# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
Total_num_schools = school_data_complete['school_name'].nunique()
#Total_num_schools
```

```python
Total_num_students = (student_data["Student ID"].max() + 1)
#Total_num_students
```

```python
Total_budget = school_data["budget"].sum()
#Total_budget
```

```python
Average_math_score = school_data_complete["math_score"].mean()
#Average_math_score
```

```python
Average_reading_score = school_data_complete["reading_score"].mean()
#Average_reading_score
```

```python
Num_student_pass_math = (len(student_data[(student_data['math_score'] >= 70)])/Total_num_students)*100
#Num_student_pass_math
```

```python
Num_student_pass_reading = (len(student_data[student_data['reading_score'] >= 70])/Total_num_students)*100

#Num_student_pass_reading
```

```python
passing_rate = (Num_student_pass_math + Num_student_pass_reading)/2
#passing_rate
```

```python
school_data_agg = pd.DataFrame(
    {'Total Number of Schools':[Total_num_schools],
     'Total Number of Students':[Total_num_students],
     'Total Budget':[Total_budget],
     'Average Math Score':[Average_math_score],
     'Average Reading Score':[Average_reading_score],
     '% Passing Math': [Num_student_pass_math],
     '% Passing Reading': [Num_student_pass_reading],
     'Passing Rate': [passing_rate]
    }
)
#Improve formatting before printing
school_data_agg["Total Budget"] = school_data_agg["Total Budget"].map("${0:,.0f}".format)
school_data_agg["Average Math Score"] = school_data_agg["Average Math Score"].map("{0:,.0f}".format)
school_data_agg["Average Reading Score"] = school_data_agg["Average Reading Score"].map("{0:,.0f}".format)
school_data_agg["% Passing Math"] = school_data_agg["% Passing Math"].map("{0:,.3f}%".format)
school_data_agg["% Passing Reading"] = school_data_agg["% Passing Reading"].map("{0:,.3f}%".format)
school_data_agg["Passing Rate"] = school_data_agg["Passing Rate"].map("{0:,.0f}".format)
school_data_agg
```

```python
school_types = school_data.set_index(["school_name"])["type"]
#school_types
```

```python
student_per_school = school_data_complete["school_name"].value_counts()
#student_per_school
```

```python
Budget_per_school = school_data.groupby(['school_name']).budget.sum()
#Budget_per_school
```

```python
Per_student_budget = Budget_per_school/student_per_school
#Per_student_budget
```

```python
Avg_math_score_by_school = school_data_complete.groupby(['school_name'])["math_score"].mean()
#Avg_math_score_by_school
```

```python
Avg_reading_score_by_school = school_data_complete.groupby(['school_name'])["reading_score"].mean()
#Avg_reading_score_by_school
```

```python
group_keys = (school_data_complete.school_name)
Num_student_pass_math_by_school = ((school_data_complete['math_score'] >= 70).groupby(group_keys).sum()/student_per_school)*100

#Num_student_pass_math_by_school
```

```python
#Num_student_pass_reading_by_school = (((school_data_complete['reading_score'] >= 70).groupby(group_keys).sum())/student_per_school)*100 
group_keys = (school_data_complete.school_name)
Num_student_pass_reading_by_school = ((school_data_complete['reading_score'] >= 70).groupby(group_keys).sum()/student_per_school)*100

#Num_student_pass_reading_by_school
```

```python
passing_rate_by_school = (Num_student_pass_math_by_school + Num_student_pass_reading_by_school)/2
#passing_rate_by_school
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
# Creating a new DataFrame using school name, School type, Total students
school_summary_table = pd.DataFrame({"School Type" : school_types,
                                     "Total Students": student_per_school,
                                    "Total School Budget": Budget_per_school,
                                    "Per Student Budget": Per_student_budget,
                                    "Average Math Score": Avg_math_score_by_school,
                                    "Average Reading Score": Avg_reading_score_by_school,
                                    "% Passing Math": Num_student_pass_math_by_school,
                                    "% Passing Reading": Num_student_pass_reading_by_school,
                                    "Overall Passing Rate": passing_rate_by_school})

#Improve formatting before printing
school_summary_table["Total School Budget"] = school_summary_table["Total School Budget"].map("${0:,.0f}".format)
school_summary_table["Per Student Budget"] = school_summary_table["Per Student Budget"].map("${0:,.0f}".format)
#school_summary_table["Average Math Score"] = school_summary_table["Average Math Score"].map("{0:,.0f}".format)
#school_summary_table["Average Reading Score"] = school_summary_table["Average Reading Score"].map("{0:,.0f}".format)
school_summary_table["% Passing Math"] = school_summary_table["% Passing Math"].map("{0:,.3f}%".format)
school_summary_table["% Passing Reading"] = school_summary_table["% Passing Reading"].map("{0:,.3f}%".format)
school_summary_table["Overall Passing Rate"] = school_summary_table["Overall Passing Rate"].map("{0:,.0f}".format)
school_summary_table
```
